```python
def fibonacci_list(n):
    a,b,l=1,1,[1,1]
    if n<3:
        return(l[:n])
    else:
        for i in range(n-2):
            a,b=a+b,a
            l.append(a)
        return(l)
print(fibonacci_list(1))
print(fibonacci_list(2))
print(fibonacci_list(5))
print(fibonacci_list(10))
def plotfibonacci(n):    
    from matplotlib import pyplot as plt
    x=list(range(1,n+1))
    y=fibonacci_list(n)
    plt.plot(x, y)
    plt.title("Graph of fibonacci numbers till nth position")
    plt.show()
plotfibonacci(10)
```

```python
def dictsum(d1,d2):
    d3={}
    d1k=list(d1.keys())
    d2k=list(d2.keys())
    for i in d1k:
        if i in d2k:
            d3.update({i:d1[i]+d2[i]})    #updating common elements into d3
            del d2[i]
        else:
            d3.update({i:d1[i]})     #updating the d1 elements in d3
    d3.update(d2)     #updating the d2 elements in d3
    print(d3)
dictsum({'a': 10, 'b': 20},{'b': 5, 'c': 15})
```

```python
def string(a):
    b=''
    for i in a:
        b+=i.lower()
    print(b)
    clb,clbd=list(b),{}     #clb -> character list of b , clbd -> clb dictionary
    for i in clb:
        clbd.update({i:clb.count(i)})
    print(clbd)
    x=list(clbd.keys())
    y=list(clbd.values())
    import matplotlib.pyplot as plt
    plt.bar(x,y)
    plt.xlabel('Characters')
    plt.ylabel('Frequency')
    plt.title('Character Frequency')
    plt.show()
string('Probability')
```

```python
def remdups(l):     #remove duplicates
    u,i=[],0
    while i<len(l):
        if l[i] in u:
            l.pop(i)
        else:
            u.append(l[i])
            i+=1     #i only has to be inc when a element is no removed
    print(l)
    print('Mean of the list is:',sum(l)/len(l))
    l.sort()
    ll=len(l)
    if ll%2==0:
        print('Median of the list:',(l[ll//2]+l[ll//2-1])/2)
    else:
        print('Median of the list:',l[ll//2])
remdups([1,3,3,3,2,2,5,5,5,5,5,4,4,4,4])
```

```python
def stropr(l):     #string operations
    ls,svc,d='',0,[]     #ls -> longest string, svc -> start with vowel count, d -> list of length of strings
    for i in l:
        if len(i)>len(ls):     #for longest string
            ls=i
        if i[0] in ['a','e','i','o','u']:     #for vowel starting count
            svc+=1
        d.append(len(i))     #for histogram
    print('The longest string is:',ls)
    print('No. of strings starting with vowels:',svc)
    import matplotlib.pyplot as plt
    plt.hist(d)
    plt.show()
stropr(['Given','a','list','of','strings','write','a','Python','function','to'])
```

```python
def anagramfinder(l,w):
    al,w=[],sorted(w)     # al -> answer list
    for i in l:
        if sorted(i)==w:     #if i were to be a permutation of the input word, their sorted() values would be same
            al.append(i)
    print(' '.join(al))
anagramfinder(["enlist", "google", "inlets", "banana"],'listen')
```

```python
def sets(l1,l2):
    x,y=set(l1),set(l2)
    u=x
    u.update(y)     #union
    x,y=set(l1),set(l2)
    s=(x-y)     #only set 1
    s.update(y-x)     #symmetric difference, =only set 1 U only set 2
    d=u-s     # intersection = union - symmetric difference
    print(d)
    print(u)
    print(s)
    l1,l2=set(l1),set(l2)
    from matplotlib_venn import venn2  
    from matplotlib import pyplot as plt 
    venn2(subsets = (len(l1-l2),len(d),len(l2-l1)), set_labels = ('Set 1', 'Set 2')) 
    plt.show()
sets([1,2,3,4,5],[3,4,5,6])
```

As lists were asked to be taken as input, list data types were used in the function parameters. But, since union, intersection, and Symmetric difference make sense only on sets, the lists are explicitly converted into sets in the function. Also, as .update() updates the value of the set itself, it might affect the upcoming operations, so x and y have been used repetitively by setting them to set(l1) and set(l2) after every operation to avoid discrepancies.

```python
f=open("test.txt",'w+')
f.write("Write a program to Read a file and count number of words in it. Find the most frequent word. Visualize the word frequencies using a bar chart.")
f.seek(0)
fd=f.read()
fdl=fd.split(' ')
print('The number of words is:',len(fdl))
word,count=[],[]
for i in fdl:
    if i not in word:
        word.append(i)
        count.append(1)
    else:
        count[word.index(i)]+=1
print('The word occuring the most number of times is:',word[count.index(max(count))])
import matplotlib.pyplot as plt
plt.bar(word,count)
plt.show()
```
